```python
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt
import math

from helpers import rdivide,ldivide, kalmanFilter
```

### Load the data

```python
train6 = sio.loadmat('data/training6.mat')
sol6 = sio.loadmat('data/solution6.mat')
```

```python
ball = train6['ball']
rgb = train6['rgb']
t = train6['t']
predictions = sol6['predictions']
```

### Visualize the data

```python
plt.figure(figsize=(10,5))
plt.plot(ball[0,:],ball[1,:],'bo-')
plt.plot(ball[0,-1], ball[1,-1],'s', markersize=10, markeredgecolor=[0,.5,0], markerfacecolor='r')
plt.plot(ball[0,0], ball[1,0],'s', markersize=10, markeredgecolor=[0,.5,0], markerfacecolor='g')
plt.xlabel('X (meters)')
plt.ylabel('Y (meters)')
plt.axis('equal')
```

### Estimating the ball tracking states

```python
state = np.array([0,0,0,0])
last_t, N = -1,91
myPredictions = np.zeros((2, N))
param = {}
for i in range(N):
    px, py, state, param = kalmanFilter(t[0, i],
                                       ball[0, i],
                                       ball[1,i],
                                       state,param,
                                       last_t)
    last_t = t[0, i]
    myPredictions[0, i] = px
    myPredictions[1,i] = py
```

```python
plt.figure(figsize = (10, 5))
plt.plot(ball[0,:], ball[1,:], 'bo-')
plt.plot(ball[0,-1], ball[1,-1], 's', markersize=10, markeredgecolor = [0,.5,0], markerfacecolor = 'r')
plt.plot(ball[0,0], ball[1,0], 's', markersize=10, markeredgecolor = [0,.5,0], markerfacecolor='g')
plt.plot(myPredictions[0,:],myPredictions[1,:],'k+-')
plt.xlabel('X (meters)')
plt.ylabel('Y (meters)')
plt.axis('equal')
```

### Estimating errors of prediction (calculated vs. provided)

```python
myError = myPredictions[:,:-11]-ball[:,10:-1]
myError_dist = np.sqrt(myError[0,:]**2+myError[1,:]**2)
myError_mean = np.mean(myError_dist)
```

```python
plt.plot(myError_dist)
plt.title('Prediction Error over Time')
plt.xlabel('Frame')
plt.ylabel('Error (meters)')
print('Your Prediction: '+str(myError_mean))
```
